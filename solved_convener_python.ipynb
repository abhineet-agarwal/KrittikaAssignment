# Python Assignment - Working With Events

In this question, you are given 4 text files. 
These files contain timestamps, which correspond to the times of some 'events'. 

Your first task is to find out how many events are present in each file.

Consider only those events which have a positive value for timestamp


```python
import numpy as np
def convert(f):
    v=f.read().split(',')
    arr=[]
    for i in v:
        arr.append(float(i))
    return np.array(arr)
def count(v, i):
    n=np.count_nonzero(v>0)
    print(f"Number of events in file event_Q{i} is {n}")

with open('event_Q1.txt','r') as f1, open('event_Q2.txt','r') as f2, open('event_Q3.txt','r') as f3, open('event_Q4.txt','r') as f4:  
    v1=convert(f1)
    v2=convert(f2)
    v3=convert(f3)
    v4=convert(f4) 
    count(v1,1)
    count(v2,2)
    count(v3,3)
    count(v4,4)
```

After this, for each file, make a plot of events per second in that file.
You dont have to find the average rate, instead for each second 0-1, 1-2, 2-3, .. 54-55, find the number of events in each second and plot it with respect to time.

Plot these 4 separately, as well as in a single overlapping plot.
Make sure to give proper titles, labels and legends.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def convert(f):
    v=f.read().split(',')
    arr=[]
    for i in v:
        arr.append(float(i))
    return np.array(arr)
def yaxis(v):
    y=[0]*(int(v.max())+1)
    for i in v:
        n=math.floor(i)
        y[n]=y[n]+1
    return y

    
with open('event_Q1.txt','r') as f1, open('event_Q2.txt','r') as f2, open('event_Q3.txt','r') as f3, open('event_Q4.txt','r') as f4:  
    v1=convert(f1)
    v2=convert(f2)
    v3=convert(f3)
    v4=convert(f4) 
    y1=yaxis(v1)
    y2=yaxis(v2)
    y3=yaxis(v3)
    y4=yaxis(v4)
    figure, axis=plt.subplots(2,2)
    axis[1,0].plot(np.arange(0.5,len(y1)+0.5,1),y1)
    axis[1,0].set_title("Q1")
    axis[1,1].plot(np.arange(0.5,len(y2)+0.5,1),y2)
    axis[1,1].set_title("Q2")
    axis[0,1].plot(np.arange(0.5,len(y3)+0.5,1),y3)
    axis[0,1].set_title("Q3")
    axis[0,0].plot(np.arange(0.5,len(y4)+0.5,1),y4)
    axis[0,0].set_title("Q4")
    figure.text(0.5, 0.04, 'time', ha='center')
    figure.text(0.04, 0.5, 'frequency', va='center', rotation='vertical')
    plt.show()
    
    plt.plot(np.arange(0.5,len(y1)+0.5,1),y1, color='r', label='Q1')
    plt.plot(np.arange(0.5,len(y2)+0.5,1),y2, color='g', label='Q2')
    plt.plot(np.arange(0.5,len(y3)+0.5,1),y3, color='b', label='Q3')
    plt.plot(np.arange(0.5,len(y4)+0.5,1),y4, color='y', label='Q4')
    plt.xlabel("time")
    plt.ylabel("frequency")
    plt.title("Overlapped")
    plt.legend()
    plt.show()
    
    
```

Now for the final part of this question. We can see that the event frequency varies with time. Count how many bins are there, where atleast 2 of the files have more than 140 events in that bin, simultaneously.

```python
import numpy as np
import math
def convert(f):
    v=f.read().split(',')
    arr=[]
    for i in v:
        arr.append(float(i))
    return np.array(arr)
with open('event_Q1.txt','r') as f1, open('event_Q2.txt','r') as f2, open('event_Q3.txt','r') as f3, open('event_Q4.txt','r') as f4:  
    v1=convert(f1)
    v2=convert(f2)
    v3=convert(f3)
    v4=convert(f4) 
    files=[v1,v2,v3,v4]
    min_length = min(len(file) for file in files)
    count = 0
    for bin_index in range(min_length):
        bin_counts = [file[bin_index] for file in files]
        if sum(count > 140 for count in bin_counts) >= 2:
            count += 1

    
    print("Number of bins with at least two files having more than 140 events:", count)
    
```
